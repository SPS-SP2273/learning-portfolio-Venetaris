## Exercise 5

```python
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]
x_1_a = np.array(x_1)
x_2_a = np.array(x_2)

# Finding Means
mu_x = np.mean(x_1_a)
mu_y = np.mean(x_2_a)
print(mu_x, mu_y)
# x_1 has a mean of 38.5, x_2 has a mean of 302.5

# Finding standard deviation
sigma_x = np.std(x_1_a)
sigma_y = np.std(x_2_a)
print(sigma_x, sigma_y)
# x_1 has a mean of 32.420, x_2 has a mean of 326.089 (to 3 d.p.)

# Finding Correlation Coefficient without np.corrcoef somehow
CC = (sum((x_1_a - mu_x)*(x_2_a - mu_y))/np.sqrt((sigma_x**2)*(sigma_y**2)))/10
print(CC)
# Correlation Coefficient of 0.987 (to 3 s.f.)

# Comparing results to np.corrcoef
# Returns an array for x_1 w.r.t. x_2 and x_2 w.r.t. x_1 (Is it ever possible for there to be a difference?)
np_CC = np.corrcoef(x_1_a, x_2_a)
np.isclose(np_CC[0, 1], CC)         # Is True so good job to me
```

For reference
$$
\mu = \dfrac {1}{n} \sum_{n}^{i = 1} x_i
$$
$$
\sigma = \sqrt{ \dfrac {1}{n} \sum_{i = 1}^{n} (x_i - \mu)^2}
$$
$$
r = \dfrac {\sum_{i = 1}^{n} (x_i - \mu _X)(y_i - \mu _ Y)}{\sqrt {\sum_{i = 1}^{n} (x_i - \mu _X)^2 \sum_{i = 1}^{n} (y_i - \mu _Y)^2} }
$$
