# Exercise 1: Fill in the blanks

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,                # Ooooo so errors can be the same
             color='darkred', linestyle=':',        # Dark red just to mess with us eh?
             marker='o' , markerfacecolor='y',
             ecolor= 'black', capsize=5,            # Ecolor for error bar color, capsize for the size of the cap (not the nautical term)
             label='Y-values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

# Exercise 2: A cunning plot

For reference
$$
y = e^{-x/10}  \sin(x)
$$
Criteria:
- The x and y axes have labels of fontsize of 15                                &#10003;
- The x and y axes labels are in the color, darkorange.                         &#10003;
- There is a grid with an opacity of 25%.                                       &#10003; 
- Indicate the equation of the function in the title with a font size of 20.    &#10003;    
- Change the limit (using ylim) to [-1,1]                                       &#10003;

```python
x = np.linspace(0,50, 202)
y = np.exp(-x/10)*np.sin(x)
# help(np.exp)                                        Making sure I guessed the right function
plt.plot(x,y)
plt.xlabel('x', c = 'darkorange', fontsize = 15)    # Organising axis labels
plt.ylabel('y', c = 'darkorange', fontsize = 15)
plt.ylim([-1,1])
plt.grid(alpha = 0.25)
plt.title(r'$y = e ^{-x/10} \sin(x)$', fontsize = 20)
```
