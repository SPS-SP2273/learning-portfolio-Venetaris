# Task 2: Renaming Automated

```python
def rename_my_file(old_file_name):
    new_name = f"{dict[old_file_name[0:3]]}{old_file_name[3:]}"
    return new_name
```

# Task 3: Testing out the function

```python
file_names =    ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
                 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
                'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
                'jan-01_tokyo.txt']
# Setting up file names

for i in range(len(file_names)):
    old_n = file_names[i]
    file_names[i] = rename_my_file(file_names[i])
    new_n = file_names[i]
    print(f"{old_n} ----> {new_n}")
print(file_names)

# It just works
```

# Task 4: Now in practice

```python
FN = glob.glob('*')                             # For file names
print(FN)
for i in range(len(FN)):
    try:
        old_n = FN[i]
        FN[i] = rename_my_file(FN[i])
        new_n = FN[i]
        os.rename(old_n, new_n)
        print(f"{old_n} ----> {new_n}")
    except KeyError:
        print("File has most likely already been changed")
        continue

# IT WORKS, IT LIVES
# Apologies for the horrifically long renaming message needed to make sure it did not break
```

# Task 5: Creating Folders

```python
cities =    ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
             'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
for c in cities:
    if not os.path.exists(c):
        os.mkdir(c)
        print(f"Making folder {c}")
    else:
        print(f"Folder {c} has already been created")
```

# Task 6:

```python
C_R = glob.glob('*_cairo*')     # For Cairo Readings
print(C_R)                      # Oh so thats how it works
for r in C_R:
    ptd = os.path.join('Cairo')
    shutil.copy(r, ptd)
    print(f"Copied file {r} to {ptd}")

# WHOOOO FIRST TRY BABY IM ON FIRE
# Explanation for the excitement: I'm this close to putting my fist through the screen it took me 1 hour to get through the first 2 tasks
```
