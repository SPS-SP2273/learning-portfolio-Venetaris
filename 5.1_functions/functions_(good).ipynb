## 1.2 try-except

```python
number=input("Give me a number and I will calculate its square.")
square=int(number)**2              # Convert English to number
print(f'The square of {number} is {square}!')
```

```python
try:                                                                    # Do something
    number=input("Give me a number and I will calculate its square.")
    square=int(number)**2
    print(f'The square of {number} is {square}!')
except:                                                                 # In case things don't work 
    print(f"Oh oh! I cannot square {number}!")
```

Wow predicted that in lecture.  
Useful for trying stuff again when conditions change

```python
# try: tries to execute function
# except: if exception returned on any part of the function, abort previous function and move on
# In other words, controlled evacuation from the code
```

## 2.2 Docstrings

```python
def side_by_side(a, b, c=42):
    '''
    A test function to demonstrate how 
    positional, keyword and default arguments 
    work.
    '''
    return f'{a: 2d}|{b: 2d}|{c: 2d}'
help(side_by_side)
```

```python
def tst_Fun(a,b,c): # Ok works now docstring needs to be before the actual function
    '''
    Testing if this works now 
    '''
    A = sum(a,b,c)
    return A 
help(tst_Fun)
```
