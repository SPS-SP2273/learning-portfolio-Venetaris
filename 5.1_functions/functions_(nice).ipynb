## 2.1 *args & **kwarg

### *args

```python
def multiply(x, y):
    return x * y

numbers = [1, 2]
multiply(*numbers) #Vs
# multiply(numbers)

# * helps to tell python that the numbers are arguments, so going back to sum earlier
def add(*argums): # Allows for variable number of arguments
    result = 0
    for num in argums:
        result += num
    return result
add(*[1,2,3,4,5])
```

### **kwargs

```python
# Keyword arguments can be passed with a dictionary
def add_powers(numbers, power):
    result = 0                        # sets results variable
    for number in numbers:
        result += number**power       # loops the addition of powers

    return result

#Using **  
kwargs = {'numbers': [2,5,1,3], 'power': 3}
add_powers(**kwargs)
```
