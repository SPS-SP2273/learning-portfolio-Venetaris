## 1.1 Named Functions

### Named functions that return

```python
def func(arg):
    if arg == "Sad":
        return 'Cheer up man'
    elif arg == "Happy":
        return "Good for you, back to work!"
    else:
        return "What"
func("Happy")
```

```python
# Trying out a 5 number summary with np
import numpy as np
def FNS(Numbers):
    num_arr = np.array(Numbers)
    my_min = num_arr.min
    my_Q1, my_med, my_Q3 = np.quantile(num_arr, [0.25,0.50,0.75])
    my_max = num_arr.max
    return my_min, my_Q1, my_med, my_Q3, my_max
FNS([1,2,3,4,5,6,7,8,9])
```

## 1.2 Anonymous functions

```python
numbers=[[9, 0, -10],
         [8, 1, -11],
         [7, 2, -12],
         [6, 3, -13],
         [5, 4, -14],
         [4, 5, -15],
         [3, 6, -16],
         [2, 7, -17],
         [1, 8, -18],
         [0, 9, -19]]
sorted(numbers, key = lambda y: y[1])
# Lambda accepts argument y, y is replaced by each element in numbers in this case
# Best used as a function inside other functions, so can modify existing ones to fit use case
```
