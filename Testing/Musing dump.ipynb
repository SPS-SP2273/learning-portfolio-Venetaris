## Excess stuff

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation as ani

t = 0
x = np.linspace(0, 2*np.pi, 50)
y0 = np.sin(x- 0.2*t)
y1 = np.cos(x- 0.4*t)

fig, ax = plt.subplots()

my_plot1, = ax.plot(x,y0,'b')
my_plot2, = ax.plot(x,y1,'r')

def update(t):
    y0 = np.sin(x- 0.2*t)
    y1 = np.cos(x- 0.4*t)
    my_plot1.set_data(x,y0)
    my_plot2.set_data(x,y1)
    ax.set_title(f'time = {t}')    

mate = ani.FuncAnimation(
    fig = fig, func = update, frames = 60, interval = 1
)
mate.save('ani.gif')
```
