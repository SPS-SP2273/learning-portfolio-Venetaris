### Classes and Object-Oriented Programming (OOP)

```python
class CParticle1D:              # 1 dimensional class of object
    mass=None
    position=None
    radius=None                             # Class variables

    def __init__(self, position) -> None:   # Object variables
        self.position = position
        pass


particle_0=CParticle1D()
particle_0.mass=10        
particle_0.position=.5         
particle_0.radius=.01 

print(particle_0)               # Shows object type as well as the location

print(f"{particle_0.mass}, {particle_0.position}, {particle_0.radius}") # Need to extract results from this
```
