## Exercise 3: Words
| #  | **Term**     | **Description**                                                                                                           |
|----|--------------|---------------------------------------------------------------------------------------------------------------------------|
| 1  | Function     | Operator that does something with the variable in its (), needs to be called with ().                                     |
| 2  | Arguments    | A variable (string, numeric, etc) that is accepted by a function to perform.                                              |
| 3  | Comments     | Denoted by a # in front, used to leave notes and comments for organisation while not messing with the code.               |
| 4  | Indentations | Separates blocks of code, like conditions from executed code, and makes python more readable.                             |
| 5  | Packages     | Sets of pre-built functions not available in base Python made by other users for free use after importing.                |
| 6  | .            | Used as an indicator of ownership, aids in calling functions or attributes of packages and variables.                     |
| 7  | :            | Divides conditions from the functions they perform.                                                                       |
| 8  | []           | Creates a one-dimensional list of variables.                                                                              |
| 9  | [[],[]]      | Creates a 2 dimensional list.                                                                                             |
| 10 | {}           | Stores data in a dictionary.                                                                                              |
