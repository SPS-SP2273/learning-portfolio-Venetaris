```python
pi = 3.141592653589793
print(f"The value of pi to 2 decimal places is {pi:.2f}")
print(f"The value of pi to 3 decimal places is {pi:.3f}")
print(f"The value of pi to 4 decimal places is {pi:.4f}")
```

```python
A = 0.99
B = 0.59
O = 1.29
print(f"{'Apple':<10} ${A:>5.2f} \n{'Banana':<10} ${B:>5.2f} \n{'Orange':<10} ${O:>5.2f} \n********** $*****")
#is there a better way to do this apart from just hardcoding in the spaces?
```

```python
# Trying out new coding method for above
fruit = ['Apple', 'Banana', 'Orange', '**********']
price = [0.99, 0.59, 1.29, "*****"]
for i in range(len(fruit)):
    print(f'{fruit[i]:<10} ${price[i]:>5}') # ah yeah f strings make sense
# altered the code slightly to get the bottom row, pretty cool, incorporated the f-string formatting into above
```

## Exercise 2: Grade measurement

```python
x = int(input("Enter the students score"))
print(x)
Grade = ""

if x >= 70:
    Grade = "A"
elif x >= 50:
    Grade ="B"
elif x >= 35:
    Grade = "C"
elif x >= 0:
    Grade = "Fail"
else:
    Grade = "Unrecognized, check score and re-enter"
print(f"The student's letter grade is: {Grade}") #Yey
```

```python
# Fixed code
def factorial(n):
    if n < 0:
        return "Invalid input"
    elif n == 0:
        return 1
    else:
        result = 1
        for number in range(1, n+1):
            result *= number
        return result


print(factorial(5))   # Testing, expected output: 120
print(factorial(-1))  # Testing, expected output: Invalid input
```
