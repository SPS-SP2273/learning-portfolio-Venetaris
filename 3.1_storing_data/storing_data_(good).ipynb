# 1 Subsetting: Indexing and Slicing

## 1.1 Lists & Arrays in 1D | Subsetting & Indexing

```python
# In 1D, subsetting same for both array and list, practiced here with array
import numpy as np
arr = np.array(['a0','b1','c2','d3','e4','f5']) 
arr[3:5] # gives 2, exclusive of 5, RMB THIS BIT
arr[0:5:2] # gives elements in steps of 2
arr[::-1] #reverses list, gives all elements in steps of -1
```

## 1.2 Arrays only | Subsetting by masking

```python
m_mask = arr == 'a0'
print(m_mask)
arr[m_mask]
# np.logical_and exists  
```

## 1.3 Lists & Arrays in 2D | Indexing & Slicing

```python
import numpy as np
py_list_2d = [[1, "A"], [2, "B"], [3, "C"], [4, "D"],
              [5, "E"], [6, "F"], [7, "G"], [8, "H"],
              [9, "I"], [10, "J"]]

np_array_2d = np.array(py_list_2d)
print(py_list_2d[3])
np_array_2d[3]
np_array_2d[3, 0]  # Gives 1st element of 4th row
# list and element work slightly differently
# lists:
py_list_2d[:3][0]  # Gives 1st element of the list subsetted from [:3]
# array:
np_array_2d[:3, 1]  # Gives second element of indexes 0 - 3, but has a chance of converting all to strings even if you want floats
```

## 1.4 Growing lists

```python
# Arrays converted from list to vice versa, lists are much easier to grow and stretch.
x = [5, 1, 5]*3
print(x)
# .append() is most efficient
# .extend() can add entire new elements of lists within lists
# therefore use .append() for single elements and .extend() for other lists to extend one list 
[1] + [2,3,4,5,6] # lists can be directly added to one another
```
