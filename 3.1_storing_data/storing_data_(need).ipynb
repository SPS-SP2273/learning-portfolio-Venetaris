# 1 Lists, Arrays & Dictionaries

## 1.1 Let’s compare

```python
# Creating python lists
p_list_character = ["Oltyx", "Trazyn", "Azhek", "Mephiston"]
p_list_faction = ["Necrons", "Necrons", "Thousand Sons", "Blood Angels"]

# Creating numpy arrays
np_array_char = np.array(["Oltyx", "Trazyn", "Azhek", "Mephiston"])
np_array_fact = np.array(["Necrons", "Necrons", "Thousand Sons", "Blood Angels"])

#Creating dictionary
g_dict_info = {
    "Oltyx":"Necrons",
    "Trazyn": "Necrons",
    "Azhek":"Thousand Sons",
    "Mephiston": "Blood Angels"
}
```

## 1.2 Accessing data from a list (or array)

```python
print(p_list_faction[-1])
print(np_array_char[3]) #rmb negative indexing
```

## 1.3 Accessing data from a dictionary

```python
print(g_dict_info["Oltyx"])
print(g_dict_info.values())
```

## 2.2 Arrays are fussy about type

```python
python_l = [1, "A", True]
print(python_l)
num_1 = np.array([python_l])
print(num_1) # Lowest common denominator is 'str'
```

## 2.3 Adding a number

```python
list1 = [1,2,3,4,5,6]
array1 = np.array(list1)
# list1 + 1
# print(list1) does not work
array1 += 10
print(array1)
```
