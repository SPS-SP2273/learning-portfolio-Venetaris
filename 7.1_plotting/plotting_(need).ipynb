# 1 A simple plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')   # First plotting command, identify variable in order of x then y, color and line
plt.xlabel('x-values')                      # X labelling
plt.ylabel('y-values')                      # Y labelling
plt.title('X vs Y')                         # Titling
plt.grid(alpha=.25)                         # Adding grid with transparency
plt.legend(loc='upper left')                # Adding location
```

```python
# Improving following graph with above aspects in mind
x = np.linspace(0,10,100)
y = x
y_2 = x **2
plt.plot(x,y, label = "y = x")                              # Label is simply to describe the line, and can be shown with the legend function
plt.plot(x,y_2, 'r', ls = ':', lw = 10, label = "y = x^2")  # Plotting after another plot will add the plot on to the same axis. Ls is same as linestyle, lw is linewidth
plt.xlabel("x-values(units)")
plt.ylabel("y-values(units)")
plt.title("x against y")
plt.grid(alpha = 0.9)
plt.legend(loc = "upper left")
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = (.2,.3,.4,.5,.6,.7,.8), (.1,.2,.3,.4,.5,.6,.7)       # Can input list with index
                                                                        # Errors need their own list 

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')                     # How to insert error bars,
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
plt.style.use('fivethirtyeight')
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly)
plt.ylim([-3, 3])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
