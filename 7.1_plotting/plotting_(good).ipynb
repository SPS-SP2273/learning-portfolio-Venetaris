# 2 Comparing the two ‘dialects’

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()                        # Uses variable 1 as 1 axis, and variable 2 as another
```

### OO Interface

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharey= True)        # Matplotlib creates both a figure and an axis
ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

for a in ax:
    a.legend(loc = 'upper left')
    a.grid(alpha =.25)
ax[0].set_title('Cos(x)')   # OO interface uses set_ for labels, pyplot uses direct no.
ax[1].set_title('Sin(x)')
plt.tight_layout()          # oooo cool
```

# 3 What is OO ax, really?

```python
# In essence, subplots creates a small table space that the plots occupy 
fig, ax = plt.subplots(ncols=1, nrows=2)
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey= 'row', sharex= 'col')

# Instead of indexing, ax can be split up into two spaces 
tan_x = np.tan(x)
# For 2x2

print(ax)
# Ax has become 2 dimensional

# Flattening it is possible
ax = ax.flatten()
print(ax)
ax[0].plot(x, cos_x, label= 'Cos(x)')
ax[1].plot(x,sin_x,label='Sin(x)')
ax[2].plot(x,tan_x,label='Tan(x)')
ax[3].plot(x,x, label = 'y = x')
```
