```python
i = [5,6,7,8,9]
for n, number in enumerate(i):
    print(f"The number at index {n} is {number}")
```

## 1.3 for with range

```python
x = 10
y = 21
z = 2
for i in range(x,y,z):
    print(i)
```

```python
numbers = [6,7,8,9]
Gods = ["Slaanesh", "Nurgle", "Khorne", "Tzeentch"]
for i in range(len(numbers)):
    n = numbers[i]
    g = Gods[i]
    print(f"The Chaos God {g}'s number is {n}")
    if g == "Khorne":
        print("BLOOD FOR THE BLOOD GOD, SKULLS FOR THE SKULL THRONE!")
```

# 2 while

```python
interest = 0.02 # Per annum
loan = 40000
payment = 2500 # Per month
m = 0          # Number of months
# How many months will it take to pay off the loan?
while loan >= 0:
    loan = loan - payment
    loan = loan*(1 + interest/12)
    m += 1
print(m) # It will take 17 months
```

In above case, need to find how many loops there are to be made, so while is best if iterations are finite
