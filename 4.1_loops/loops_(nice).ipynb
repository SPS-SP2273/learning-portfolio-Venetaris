# 1 There is more to list comprehension

```python
[[a,b] for a in range(5) for b in ['A', 'B', 'C'] if a % 2 != 0] # Conditions added to list of lists
```

# 2 Zipping a dictionary

```python
# Zip is super useful for taking lists and conversion into a dictionary
p_list_character = ["Orikan", "Szarekh", "Amon", "Albinius"]
p_list_faction = ["Necrons", "Necrons", "Thousand Sons", "Blood Angels"]
g_dict_zipped = dict(zip(p_list_character, p_list_faction))
print(g_dict_zipped)
```

# 3 for and while has an else

```python
for i in [0,1,2,-3,4,5,6]:
    if i < 0:
        print('there are negative numbers in the list')
        break
else:
    print('There are no negative numbers in the list') # Only executed if the loop is broken
```
