# 1 Interrupting the flow

```python
for power in range(5):
    number = 10**power
    if number > 5000:
        break           #Break is used to stop a loop from continuing further
    print(power, number) 
```

```python
# Continue practice
for i in range(10):
    no = 5*i
    if i == 5:
        continue
    else:
        print(no) # Best to add conditions to either break or continue, one is terminate, the other is skip over
```

## 2.2 List comprehension with conditions

```python
# Very different from R so take note 
[i*25 for i in range(3)] # Creates list of multiples of 25
print([i*2 for i in range(3)])
[i for i in range(5) if i % 2 == 0] # Conditions add flexibility
```

# 3 Other useful stuff

## 3.1 for with unpacking

```python
x,y,z = [1,2,3]
print(x)
# In other words:
vector = [[2,2**2],[3,3**2],[4,4**2]]
for a,b in vector:
    print(f"the square of {a} is {b}")
```

## 3.2 for with zip()

```python
p_list_character = ["Oltyx", "Trazyn", "Azhek", "Mephiston"]
p_list_faction = ["Necrons", "Necrons", "Thousand Sons", "Blood Angels"] # taking from storing data (need)
for a,b in zip(p_list_character, p_list_faction):
    print(f"The character {a} belongs to the {b}")
```

## 3.3 for with dictionaries

```python
g_dict_info = {
    "Oltyx":"Necrons",
    "Trazyn": "Necrons",
    "Azhek":"Thousand Sons",
    "Mephiston": "Blood Angels"
}
g_dict_info.items() # Returns key and corresponding value in the form of a list
for k,v in g_dict_info.items():
    print(f"{k} belongs to the {v}")
g_dict_info.keys() # Returns only the keys, just rmb for the future.
```
