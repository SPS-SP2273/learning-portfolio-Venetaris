# Exercise 1: Variable temperature converter
For reference (again):
$$
F = (C * \dfrac {9}{5}) + 32
$$
$$
K = C + 273.15
$$

```python
def convert_celsius(temperature_celsius, target_scale = 'Fahrenheit'):     # Setting up args
    if target_scale == 'Fahrenheit':
        temp_fahrenheit = (temperature_celsius*9/5) + 32 
        return f"The temperature is {temp_fahrenheit:.2f}°F"
    elif target_scale == 'Kelvin':
        return f"The temperature is {(temperature_celsius + 273.15):.2f} °K"      # Seeing if only one line is needed
    else:
        return "Target scale not recognized"

# Testing
print(convert_celsius(0))
print(convert_celsius(50))
print(convert_celsius(50, target_scale = 'Kelvin'))
print(convert_celsius(50, 'Kelvin'))
```

# Exercise 2: Same same but different

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale = 'Celsius'):     # Copy pasted everything to make things easier
    temp_celsius = (temperature_fahrenheit - 32)*5/9                          # Doing celsius conversion first to make Kelvin easier
    if target_scale == 'Celsius':
        return f"The temperature is {temp_celsius:.2f}°C"
    elif target_scale == 'Kelvin':
        return f"The temperature is {(temp_celsius + 273.15):.2f} °K"
    else:
        return "Target scale not recognized"

# Testing
print(convert_fahrenheit(32))
print(convert_fahrenheit(122))
print(convert_fahrenheit(50, target_scale = 'Kelvin'))
print(convert_fahrenheit(50, 'Kelvin'))
```
