# Exercise 1: Code detective work

````python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
Name = "Batman"            # Positive case
greeting(Name)
Name = "ManBat"            # Negative case
greeting(Name)
# Current hypothesis, return breaks out of the function, when if statement is false, skips over that section and executes last
````

````python
# Minor experimenting, question is, does return break out of the function?
thing = "Batman"
def expt(arg):
    if arg == 'Batman':
        print('Hello Batman! So, nice to meet you!') # Testing if return is the problem
    print(f'Hello {arg}!')
expt(thing) 
# Yeap as expected, executes both lines
````

Answer: The ```return``` section of the function exits the function upon doing so, and does not move on to the second line if the ```if```  is true. The ```if``` function is skipped upon being False, does not return anything, allowing the function to move on to the next line. 

# Exercise 2: BMI

````python
def calculate_bmi(weight, height): 
    BMI = weight/(height**2)
    if BMI >= 30:
        c = "Obese"                # c stands for Category
    elif BMI >= 25:
        c = "Overweight"
    elif BMI >= 18.5:
        c = "Normal weight"
    elif BMI >= 10:                 # Basically the average lowest survivable BMI
        c = "Underweight"
    else:
        c = "Not Alive, check the patient or your math" 
    return f"The patient has a BMI of {BMI:.2f} and is {c}"

print(calculate_bmi(68, 1.72)) # For myself
calculate_bmi(5, 1.80)
````

# Exercise 3: ! in code

````python
def factorio(n):                                                # Chose math game for math exercise
    if n < 0:
        return "The number is negative, please don't do that"   # Avoiding negative numbers 'gracefully'
    answer = 1                                                  # Create initial variable
    for i in range(n):
        answer *= (i + 1)                                       # Countering the 0 indexing
    return answer
print(factorio(-10))                                            # Testing negative numbers
factorio(5)                                                    # Testing if works as intended

# It does nice
````
