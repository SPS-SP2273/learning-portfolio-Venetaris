## 1 Quadratic Equations

### Introduction


![alt text](Polynomialdeg2.svg.png)

Image from the ![Wikipedia page on Quadratic equations](https://en.wikipedia.org/wiki/Quadratic_equation)

The general form of a quadratic equation is

$$
ax^2 + bx + c = 0
$$


### Solutions


Provided $x \ne 0 $ we can use an elementary algebraic method called ![completing the square](https://en.wikipedia.org/wiki/Completing_the_square) to show that a quadratic equation has the following solution:

$$
x = \dfrac {-b \pm \sqrt (b^2 - 4ac)} {2a}
$$


### Discriminant

The quantity $\Delta = b^2 - 4ac $ is called the **discriminant** of the equation and decides the nature of its solutions. The table below shows the various possibilities
 | **Discriminant**          | **Roots**                            |
|---------------------------|--------------------------------------|
| $\Delta = b^2 - 4ac = 0$  | A single solution of $ -b/(2a)$      |
| $\Delta = b^2 - 4ac > 0$  | Two distinct solutions               |
| $\Delta = b^2 - 4ac < 0$  | No real solutions; both are complex. |

1. What is a commit?
    - Commit creates an updated version of the file and tracks all changes from the previous version of the file.

    - Pushes committed changes to the origin (i.e. the repository).

1. What is a pull?
    - Pulls the changes made to the files from the Git repository and updates local files to match.
