## Exercise 1: Converting °C to °F
For reference:
$$
F = (C * \dfrac {9}{5}) + 32
$$

```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]
temperatures_farenheit = [(c*9/5 + 32) for c in temperatures_celsius]
for a, b in zip(temperatures_celsius, temperatures_farenheit):
    print(f"{a}°C is {b}°F")
```

## Exercise 2: Multiplication tables

```python
Range = [1,2,3,4,5]
for i in Range:
    for j in Range:
        print(f"{i}: {i*j}", end='\t')
    print(f"\n")
```

## Exercise 3: Coin flip 

```python
import numpy as np
np.random.rand(10)
for i in range(10):
    a = np.random.rand(10)
    h = sum(a > 0.5)
    print(f"Experiment {i + 1} has {h} heads")
```
