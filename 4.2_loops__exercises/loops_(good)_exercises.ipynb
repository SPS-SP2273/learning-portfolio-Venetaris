## Exercise 1: List generation

```python
l = []
for i in range(10):
    if i % 2 == 0:
        continue
    else:
        l.append(i**2)
print(l)
```

## Exercise 2: Whoops! All exercise 1

```python
[i**2 for i in range(10) if i % 2 != 0]
```

## Exercise 3: A Matter of Time

```python
%%timeit
l = []
for i in range(10):
    if i % 2 == 0:
        continue
    else:
        l.append(i**2)
# 1.34 microseconds
```

```python
%%timeit
[i**2 for i in range(10) if i % 2 != 0]
# 1.33 microseconds wait what its only barely faster, maybe its the exact same process just worded differently orrrrrr?
```

## Exercise 4: Half-life 3 copium

```python
rad_mat = 100
decay = 0.05
year = 0
while rad_mat >= 50:
    rad_mat = rad_mat - rad_mat*decay
    year = year + 11
print(year)
# 154 years till half life (we really are going to wait that long aren't we :( )
```
